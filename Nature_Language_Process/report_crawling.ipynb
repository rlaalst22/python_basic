```python
def pdfparser(pdf_list,pdf_name):
    try:
        location_pdf = 'C:\\Users\\82102\\Python_study\\my_python\\Nature_Language_Process\\report\\'
        fp = open(location_pdf+pdf_list, 'rb')
        rsrcmgr = PDFResourceManager()
        retstr = io.StringIO()
        codec = 'utf-8'
        laparams = LAParams()
        device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
        # Create a PDF interpreter object.
        interpreter = PDFPageInterpreter(rsrcmgr, device)
        # Process each page contained in the document.

        for page in PDFPage.get_pages(fp):
            interpreter.process_page(page)
            data =  retstr.getvalue()
            location_text = 'C:\\Users\\82102\\Python_study\\my_python\\Nature_Language_Process\\text'
            #                 text_path = os.path.join(REPORTS_DIR, i + '.txt')
            txt_name = '\\'+ pdf_name+'.txt'
            with open(location_text+txt_name, 'w', encoding='utf-8') as f:
                f.write(data)

    except Exception as e:
        print('텍스트 변환 오류 :',e)
        pass
```

```python
import re
from tika import parser
from os import listdir
from os.path import isfile, join
import shutil
# 폴더내 pdf파일을 txt 파일로 변경
def pdf2txt (source_folder="./report/", output_folder="./text/"):
    error_list = []
    # 지정 폴더 내 파일 목록 조회 (파일만)
    pdf_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    
    for pdf in pdf_files :
        pdf_tmp_filepath = source_folder + 'tmp.pdf'
        pdf_filepath = source_folder + pdf
        shutil.copyfile(pdf_filepath,pdf_tmp_filepath)



        # pdf 파일을 text로 변환
        print(pdf_filepath)
        try:
#             parsedPDF = parser.from_file(pdf_filepath)["content"]
            parsedPDF = re.sub('\n','',parser.from_file(pdf_tmp_filepath)['content'])
        except:
            print(pdf)
            error_list.append(pdf)
            pass
        output_filepath = (output_folder+pdf).replace('.pdf','.txt')
        with open(output_filepath,'w',encoding = 'utf-8') as f:
            print(output_filepath)
            f.write(parsedPDF)
            f.close()
#         with open((output_folder + pdf).replace('.pdf', '.txt'), 'w',enconding = 'utf-8') as f :
#             f.write(parsedPDF)
#             f.close() 
    return error_list
pdf2txt()
```
